### __Initial Setup__

```python
# import needed modules
import numpy as np 
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import BernoulliNB
from sklearn.metrics import accuracy_score
```

```python
# start the model
dtc = DecisionTreeClassifier()
bnb = BernoulliNB()
```

### __Import Data__

```python
# Import Train Data
df1 = pd.read_csv("training_data.csv")
df1.head()
```

```python
# Import Test Data
df2 = pd.read_csv("test_data.csv")
df2.head()
```

```python
# Drop not-needed column
df1.drop("Unnamed: 133", axis=1, inplace=True)
df1.head()
```

```python
# Check if size is compatible
df1.shape, df2.shape

# since they both have same number of columns, we know that it is compatible.
```

```python
# Extract X, Y
x_train = df1[df1.columns[:-1]]
y_train = df1[df1.columns[-1]]
x_test = df2[df2.columns[:-1]]
y_test = df2[df2.columns[-1]]

x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

### __Train Model__

```python
dtc.fit(x_train, y_train)
bnb.fit(x_train, y_train)
```

```python
yhat1 = dtc.predict(x_test)
yhat2 = bnb.predict(x_test)
scre1 = accuracy_score(y_test, yhat1)
scre2 = accuracy_score(y_test, yhat2)

print("Scores are: ", scre1, scre2)
```

## Result
We know that Naive Bayes shows better accuracy than Decision Tree. Now we can make a survey that record's patients symptoms and predict with the model.
