```python
# getting rid of excessive outliers
q1 = df["realhrwage"].quantile(0.25)
q3 = df["realhrwage"].quantile(0.75)
IQR = q3-q1
inds = df[(df["realhrwage"] > q3 + 1.5 * IQR) | (df["realhrwage"] < q1 - 1.5 * IQR)].index

df.drop(inds, inplace=True)
df.shape
```
