### Bayesian Statistics

This notebook's purpose is to calculate and graph the experimental part of the research paper.<br>
<br>
The Chicago police department wants to make a prediction concerning the percentage of
battery crimes related to all crimes reported in the city. Note that battery is when someone
physically harms another person. This data came from data.gov on the yearly crimes in
Chicago. Specifically, this is dataset from 2019. They are interested if the
prediction of the number of crimes in 2019 is more than 20%.
<br>


```python
# Importing Modules
import numpy as np
from scipy.stats import beta
from matplotlib import pyplot as plt
```

```python
# Setting Variables
labels = ["non-informative", "January", "Jan-Feb", "Jan-May"]
alpha_values = [1,48+1, 85+1, 239+1]
beta_values = [1, 222-48+1, 425-85+1, 1118-239+1]
linestyles = ['-.', '-', '--', ':']
x = np.linspace(0, 1, 1002)[1:-1]

mean = []
intervals = []
probs = []
SDs = []
```

```python
# Graphing the Plot
fig, ax = plt.subplots(figsize=(5, 3.75))

for a, b, ls in zip(alpha_values, beta_values, linestyles):
    dist = beta(a, b)

    mean.append(dist.mean())
    SDs.append(dist.std())
    intervals.append(dist.interval(0.95))
    probs.append(1 - dist.cdf(0.2))

    plt.plot(x, dist.pdf(x), ls=ls, c='black',
             label=r'$\alpha=%.1f,\ \beta=%.1f$' % (a, b))

plt.xlim(0, 1)
plt.ylim(0, 3)

plt.xlabel('$x$')
plt.ylabel(r'$p(x|\alpha,\beta)$')
plt.title('Beta Distribution')

plt.legend(loc=0)
plt.show()
```

```python
for l, a, b, m, s, i, p in zip(labels, alpha_values, beta_values, mean, SDs, intervals, probs):
    print(l)
    print("a, b:", (a, b))
    print("Mean:", m)
    print("SD:", s)
    print("Probability Interval:", i)
    print("P(X > 0.2):", p)
    print("")
```
